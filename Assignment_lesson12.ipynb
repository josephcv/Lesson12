```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Assignment 12.1 - Matplotlib excercise
# Step1: Adding color code for sex and also counting the sex attribute
# Step2: using a separate DataFrame for pie chart
# Step3: creating scatter plot using color differentiator for sex

url='http://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic=pd.read_csv(url)
countf=0
countm=0
lst=[]
for index,rows in titanic.iterrows():
    if rows['sex']=="female":
        countf+=1
        lst.append('r')
    else:
        countm+=1
        lst.append('b')
df1=pd.DataFrame({'Gender':['Female','Male'], 'Count':[countf,countm]})
titanic['Color']=lst
plt.pie(df1['Count'],labels=df1['Gender'],autopct='%.1f%%')
plt.show()

fig=plt.figure()
plt.scatter(titanic['age'],titanic['fare'],c=titanic['Color'])
plt.xlabel('Age')
plt.ylabel('Fare')
fig.suptitle('Scatter Plot')
plt.legend(titanic['sex'])
plt.show()
```
